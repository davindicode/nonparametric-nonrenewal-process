```python
x = np.random.randn(1000)

def f(x):
    return 0.25*(10.*np.sin(x) - x**3)
```

```python
x_eval = np.linspace(-4., 4., 100)
y_eval = f(x_eval)

plt.plot(x_eval, y_eval)
plt.show()
```

```python
ys = []
z = x
K = 5

for k in range(K):
    y = f(z)
    ys.append(y)
    z = y
```

```python
plt.hist(x, bins=np.linspace(-4., 4., 100))
plt.show()
```

```python
for k in range(K):
    plt.hist(ys[k], bins=np.linspace(-4., 4., 100))
    plt.show()
```
